```python
import pandas as pd
```

```python
url = "https://drive.google.com/uc?export=download&id=1bAu31pKhF-kAYq2ncCveSloH-NLg83aU"
df = pd.read_csv(url)
```

```python

# NIM LIST adalah daftar NIM Kelompok -4 
# - 1207050001 Abdullah
# - 1207050011 Alif Pandanhi Febrio
# - 1207050012 Alifia Nadiva Altafunnisa 
# - 1207050016 Arba Adhy Pamungkas
# - 1207050020 Ayuni Tia Sari
# - 1207050139 Dani Ali Kinan
# - 1207050095 Nurrohman

nim_list = [1207050001, 1207050011, 1207050012, 1207050016,1207050020,1207050095, 1207050139]
df["NIM"] = df["NIM"].astype(int)

```

```python
df_selected_nim = df[df["NIM"].isin(nim_list)]
```

```python
df_selected_nim = df_selected_nim.dropna()

```

```python
for nim in nim_list:
    df_nim = df_selected_nim[df_selected_nim["NIM"] == nim]

    print(f"\nKeterangan untuk NIM {nim}:")

    for index, row in df_nim.iterrows():
        print(
            f"NIM: {row['NIM']}, Nama Mata Kuliah: {row['Nama Mata Kuliah']}, Nilai Huruf: {row['Nilai Huruf']}"
        )
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder


df_all_nim = df.copy()
label_encoder = LabelEncoder()
df_all_nim["Nama Mata Kuliah"] = label_encoder.fit_transform(
    df_all_nim["Nama Mata Kuliah"]
)

features = df_all_nim[["Periode Semester", "Nama Mata Kuliah"]]
target = df_all_nim["NIM"]

X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
classification_report_result = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print("Classification Report:\n", classification_report_result)
```
